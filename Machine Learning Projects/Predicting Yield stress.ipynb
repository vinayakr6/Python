### Model Building

**Splitting the data into X & Y for both cases**

```python
X = df1.drop('YS (MPa)', axis = 1)
y = df1['YS (MPa)']
```

```python
X.shape, y.shape
```

**Checking for skewness**

```python
X.skew()
```

```python
# Let's plot distribution plot
X.plot(kind='density', subplots = True, layout = (4,6), sharex = False, legend= False, fontsize =1, figsize = (18,12))
plt.show()
```

```python
# Lets remove the skewness using transformation for regression
for index in X.skew().index:
    if X.skew().loc[index]>0.5:
        X[index]=np.log1p(X[index])
    if X.skew().loc[index]<-0.5:
        X[index]=np.square(X[index])
```

```python
X.skew()
```

**Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
z1 = MinMaxScaler()
X = z1.fit_transform(X)
X
```

```python
# Importing the libraries for model building
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instantiating each algorithm
lr = LinearRegression()
dtr = DecisionTreeRegressor()
knr = KNeighborsRegressor()
svr = SVR()
rfr = RandomForestRegressor()

# List of instantiated algorithms
model = [lr, dtr, knr, svr, rfr]
```

```python
# Maximum accuracy
maxAccu = 0 

# Best random state value for which accuracy is achieved 
maxRS = 0

for m in model:
    for i in range(0,70):
        x_train, x_test,y_train, y_test = train_test_split(X,y, random_state=i, test_size=0.20)
        m.fit(x_train, y_train)
        pred_train= m.predict(x_train)
        pred_test = m.predict(x_test)
        score_train = m.score(x_train, y_train)
        score_test = m.score(x_test, y_test)
        r2 = r2_score(y_test, pred_test)
        print(f'----------------At random State {i} ------------------')
        print(f"Training score is {score_train}")
        print(f"Testing score is {score_test}")
        print(f'r2 score of {m} is {r2}')
    
        if r2 > maxAccu:
            maxAccu = r2
            maxRS = i
            print(f"Maximum score of {m} is {maxAccu} at Random state {i}")
            print('\n')
    print('*'*120)
```

*Observations*
- ***Random Forest Regressor has given best results with following accuracy***
- Training score is 0.9842399197537791
- Testing score is 0.9893864519631593
- r2 score of RandomForestRegressor() is 0.9893864519631593
- Maximum score of RandomForestRegressor() is 0.9893864519631593 at Random state 13

**Cross validation of all the Algorithms for regression**

```python
# Before proceeding with the hpyer parameter tunning we will check for cross validation
from sklearn.model_selection import cross_val_score
max_cvscore = 0 
max_cv = 0
for m in model:
    print(f'---------------------- Cross Validation of {m} --------------------------')
    for j in range(2,5):
        cv_score = cross_val_score(m, X,y,cv=j)
        cv_mean = cv_score.mean()
        print(f"At cross fold {j} the cv score is {cv_mean}")
        print('\n') 
    
        if cv_mean > max_cvscore:
            max_cvscore=cv_mean
            max_cv = j
            print(f"At cross fold {j} the Maximum CV score is {max_cvscore}")
            print('\n') 
    print('*'*120)
    print('\n')
```

*Observations*
- At cross fold 3 the Maximum CV score is 0.358957403572506 for Random Forest Regressor

```python
# Training Random forest Regressor at Random state 13
x_train, x_test,y_train, y_test = train_test_split(X,y, random_state=13, test_size=0.20)
rfr.fit(x_train, y_train)
pred_train= rfr.predict(x_train)
pred_test = rfr.predict(x_test)
score_train = rfr.score(x_train, y_train)
score_test = rfr.score(x_test, y_test)
r2 = r2_score(y_test, pred_test)
print(f"Training score is {score_train}")
print(f"Testing score is {score_test}")
print(f'r2 score of {rfr} is {r2}')
```

***Hyper Parameter Tunning***

```python
from sklearn.model_selection import GridSearchCV
param = {'n_estimators':[100],'max_depth': [2,4],'criterion': ['mse','mae'],'max_features':['sqrt','log2']}

gcv1 = GridSearchCV(rfr,param,cv=3,scoring= 'accuracy')

# training
gcv1.fit(x_train, y_train)

# Getting the best parameters
print("Best parameters:", gcv1.best_params_)

print("Best Estimator:",gcv1.best_estimator_)

gcv_pred=gcv1.best_estimator_.predict(x_test)

print('Final Accuracy with Random Forest Regression:', r2_score(y_test,gcv_pred))
```

***Pickling Regression Model***

- Final Selected algorithm for deployment is Random Forest Regressor
